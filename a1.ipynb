```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
sns.set(style="whitegrid")
```

```python
df = pd.read_csv("areas.csv")
```

```python
df.head()
```

```python
S_exact = 0.25 * 3.141592653589793 + 1.25 * np.asin(0.8) - 1

plt.figure(figsize=(12, 8))
sns.lineplot(data=df, x="N", y="S_wide", label=" ")
sns.lineplot(data=df, x="N", y="S_narrow", label=" ")
plt.axhline(S_exact, linestyle="--", label=" ", color="red")
plt.xlabel("N")
plt.ylabel("")
plt.title(" ")
plt.legend()
plt.savefig('area.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.lineplot(data=df, x="N", y="rel_err_wide", label=" ")
sns.lineplot(data=df, x="N", y="rel_err_narrow", label=" ")
plt.xlabel("N")
plt.ylabel("")
plt.title("  ( )")
plt.legend()
plt.savefig('area_err.png', dpi=300, bbox_inches='tight')
plt.show()
```
